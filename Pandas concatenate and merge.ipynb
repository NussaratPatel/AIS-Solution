```python
#merging
df4 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
})
df4
```

```python
df5 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Age': [25, 30, 35, 40]
})
df5
```

```python
merged_df = pd.merge(df4, df5, on='ID', how='inner')
merged_df
```

```python
merged_df = pd.merge(df4, df5, on='ID', how='outer')
merged_df
```
